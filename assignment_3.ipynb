Create a list of random numbers and sort it in ascending and descending order. Remove the duplicates from the list and print the modified list.

```python
import random
lst = [random.randint(1,15) for x in range(15)]
print(f'The list of random number becomes: {lst}')
def ascending_sort(lst):
    sorted_lst = []
    temp_lst = lst.copy()
    while temp_lst:
        min_val = temp_lst[0]
        for x in temp_lst:
            if x < min_val:
                min_val = x
        sorted_lst.append(min_val)
        temp_lst.remove(min_val)
    return sorted_lst
print(f'The sorted list in ascending order becomes:\n {ascending_sort(lst)}')
def descending_sort(lst):
    sorted_lst = []
    temp_lst = lst.copy()
    while temp_lst:
        max_val = temp_lst[0]
        for x in temp_lst:
            if x > max_val:
                max_val = x
        sorted_lst.append(max_val)
        temp_lst.remove(max_val)
    return sorted_lst
print(f'The sorted list in descending order becomes:\n {descending_sort(lst)}')
print(f'The origional list becomes:\n {lst}')
b = []
for x in lst:
    if x not in b:
        b.append(x)
print(f'The modified list after removing the duplicate becomes:\n {b}')
```

Write a function that takes a list and returns a new list with the elements in reverse order. Print the original and reversed lists.

```python
def reverse(lst):
    count = 0
    for x in lst:
        count = count + 1
    left = 0
    right = count - 1
    while left < right:
        temp = lst[left]
        lst[left] = lst[right]
        lst[right] = temp
        left = left + 1
        right = right - 1
    return lst
lst = [x for x in range(1,6)]
print(f'The origional list becomes:\n {lst}')
print(f'The reversed list becomes:\n {reverse(lst)}')
```
