Write a program that asks the user to input numbers until they input 0. The program should print the sum of all the input numbers.

```python
total = 0
while True:
    n = eval(input('Enter numbers:'))
    print('The entered number becomes\n',n)
    if n == 0:
        break
    total = total + n
print(f'The sum of the entered number becomes {total}')
```

Write a program that prints all the numbers from 1 to 10 except 5 using a for loop and continue statement.

```python
print('The numbers are:')
for i in range(1,11):
    if i == 5:
        continue
    print(i)
```

Write a program that calculates the factorial of a number input by the user using a while loop.

```python
n = eval(input('Enter a number:'))
fact = 1
if n < 0:
    print(f'factorial of {n} is not possible')
elif n == 0:
    print(f'factoral of {n} becomes {1}')
else:
    print(f'factorial of {n} becomes:')
    i = 1
    while i <= n:
        fact = fact*i
        i = i+1
    print(fact)
        
        
        
```

Write a program that calculates the sum of the digits of a number input by the user using a while loop.

```python
n = eval(input('Enter a number:'))
total = 0
temp = n
while n > 0:
    dig = n % 10
    total = total + dig
    n = n//10
print(f'The sum of digits of {temp} becomes: {total}')
```

Write a program that checks if a number input by the user is a prime number using a for loop.

```python
n = eval(input('Entrer a number:'))
is_prime = True
if n <= 1:
    is_prime = False
else:
    for i in range(2,int(n*0.5)+1):
        if n % i == 0:
            is_prime = False
            break
if is_prime:
    print(f'{n} is a prime number')
else:
    print(f'{n} is not a prime number')    
```
