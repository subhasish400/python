## Matplotlib

Matplotlib is a powerful plotting library for python which enables the creation of static, animated and interactive visualization. It is widely used for data visualization in data science and analytics.

```python
import matplotlib.pyplot as plt
```

```python
# Write a program in python to create a basic line plot by using given data in x and y axis
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic line plot')
print(f'The basic line plot becomes:')
plt.show()
```

```python
# Write a program in python to customize the first plot
plt.plot(x,y,color='black',linestyle='--',linewidth=2,marker='o',markersize=10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Modified line plot')
print(f'The modified plot becomes:')
plt.show()
```

```python
x = [x for x in range(1,6)]
y1 = [x**2 for x in range(1,6)]
y2 = [x for x in range(1,6)]
plt.figure(figsize=(9,5))
plt.subplot(2,2,1)
plt.plot(x,y1,color='black')
plt.title('Plot-1')

plt.subplot(2,2,2)
plt.plot(y1,x,color='black')
plt.title('Plot-2')
print(f'The plot becomes:')
plt.show()
```

```python
# Write a program in python to create a bar plot
catagories = ['A','B','C','D','E']
values = [10,40,30,20,80]
plt.bar(catagories,values,color='black',alpha=0.5)
print(f'The bar plot becomes:')
plt.show()
```

Histograms are used to represent the distribution of the dataset. They divide data into bins and count the number of data points in each bin.

```python
# Write a program in python to create a histogram for given sample data
data = [1,1,2,2,2,3,3,4,4,4,5,5,5,5,5]
print(f'The histogram plot becomes:')
plt.hist(data,color='black',alpha=0.5,edgecolor='black',bins=5)
plt.show()
```

```python
# Write a program in python to create a scatter plot
x = [x for x in range(1,6)]
y = [x for x in range(2,7)]
print(f'The scatter plot becomes:')
plt.scatter(x,y,color='black',marker='o')
plt.show()
```

```python
# Write a program in python to create a pie chart
labels = ['A','B','C','D']
sizes = [30,40,20,10]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.2,0,0,0)
print(f'The pie chart becomes:')
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
plt.show()
```

Visualization based on sales data

```python
# Now we are going to load the sales data so we have
import pandas as pd     
data = pd.read_csv('sales_data.csv')
print(f'The data becomes:')
data.head(5)
```

```python
# We have to know the information about the data so we have
data.info()
```

```python
# Write a program in python to create a bar plot between total sales by product so we have
total_sales_by_product = data.groupby('Product Category')['Total Revenue'].sum()
print(f'The total sales by product becomes:\n {total_sales_by_product}')
```

```python
# Now we are going to create a bar plot so we have
total_sales_by_product.plot(kind='bar',color='black',alpha=0.5)
plt.xlabel('Products')
plt.ylabel('Total Sales')
plt.title('Total Sales by Product')
```

```python
# Write a program in python to plot sales trend over time
sales_trend = data.groupby('Date')['Total Revenue'].sum().reset_index()
sales_trend
```

```python
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
plt.show()
```
