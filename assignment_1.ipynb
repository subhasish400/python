**Question 18:** Write a Python program to reverse a string.

```python
# Your code here
s = input('Enter a string:')
count = 0
for x in s:
    count = count + 1
rev = ''
i = count - 1
while i >=0:
    rev = rev + s[i]
    i=i-1
print('The entered string becomes:',s)
print('The reversed string becomes:',rev)
```

**Question 19:** Write a Python program to check if a string is a palindrome.

```python
# Your code here
s = input('Enter a string:')
print(f'The entered string becomes: {s}')
a = list(s)
count = 0
for x in a:
    count = count+1
left = 0
right = count-1
while left < right:
    temp = a[left]
    a[left] = a[right]
    a[right] = temp
    left = left + 1
    right = right - 1
reversed = ''.join(a)
print(f'The reversed string becomes: {reversed}')
if s == reversed:
    print(f'Entered string is a palindrome string')
else:
    print(f'Entered string is not a palindrome string')
```

**Question 20:** Write a Python program to sort a list of numbers in ascending order.

```python
# Your code here
def sort(a):
    count = 0
    for x in a:
        count = count + 1
    for i in range(count-1,0,-1):
        for j in range(i):
            if a[j] > a[j+1]:
                temp = a[j]
                a[j] = a[j+1]
                a[j+1] = temp
    return a
n = eval(input('Enter total number of elements:'))
a = []
for i in range(n):
    a.append(eval(input('Enter elements:')))
print('The unsorted list becomes:',a)
print('The sorted list becomes:',sort(a))
```
