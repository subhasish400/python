```python
# Write a program in python to read data from paarticular website
import pandas as pd
def Extract(url):
    try:
        data = pd.read_html(url)
        print(f'The data becomes:')
        return data[0].head()
    except Exception as ValueError:
        print(f"It's failed to access the data because:\n {ValueError}")
        return None
url = input('Enter the url:')
# url used here: 'https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/'
Extract(url)
 
```
