```python
#import all the required packages
import numpy as np
import math # for making sine as well as cosine angle
import matplotlib.pyplot as plt #for plotting our result
# First we have to define the length of each arm
L1=1  #for first robotic arm
L2=0.5 # for second robotic arm
N_theta=10
thetastart=0
thetaend=90
# we have to convert these theta values in degree to radians,for that we have to create one empty dataset
theta1=[]
theta2=[]
y=[]
for i in range(thetastart,thetaend+1,10): # conversion of all degrees to radian
    temp=math.radians(i)
    theta1.append(temp)
    theta2.append(temp)
    y.append(i)
print(y)
print(theta1)
```

```python
x0=0 #initial position
y0=0 #initial position
figno=1
for t1 in theta1:
    for t2 in theta2:
        x1=L1*math.cos(t1)
        y1=L1*math.sin(t1)
        x2=x1+L2*math.cos(t2)
        y2=y1+L2*math.sin(t2)
        
        filename=str(figno)+'.jpg'
        figno=figno+1
        
        plt.figure()
        plt.plot([x0,x1],[y0,y1])
        plt.plot([x1,x2],[y1,y2])
        
        plt.xlim([0,2])
        plt.ylim([0,2])
        plt.savefig(filename)
        
    
```
