## Seaborn

Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code.

```python
import seaborn as sns
```

```python
# Write a program in python to load the tips dataset
tips_dataset = sns.load_dataset('tips')
tips_dataset.head()
```

```python
# Write a program in python to create a scatter plot between total_bill and tip
import matplotlib.pyplot as plt
print(f'The scatterplot becomes:')
sns.scatterplot(x='total_bill',y='tip',data=tips_dataset)
plt.title('Scatterplot between total_bill and tip')
```

```python
# Write a program in python to create a line plot between the number of people vs the total bill 
sns.lineplot(x='size',y='total_bill',data=tips_dataset)
print(f'The line plot becomes:')
plt.title('Line plot between the number of people and total_bill')
plt.show()
```

```python
# Write a program in python to create a bar plot to show the day wise average total bill for a week
sns.barplot(x='day',y='total_bill',data=tips_dataset)
print(f'The bar plot becomes:')
plt.title('Days vs Total_bill')
plt.show()
```

```python
# Write a program in python to create a box plot to show the day wise average total bill for a week to see the outliers
sns.boxplot(x='day',y='total_bill',data=tips_dataset)
print(f'The box plot becomes:')
plt.title('Days vs Total_bill')
plt.show()
```

```python
# Write a program in python to create a violin plot to show the distribution of data within each quartile for day wise average total bill for a week.
sns.violinplot(x='day',y='total_bill',data=tips_dataset)
print(f'The violin plot becomes:')
plt.title('Days vs Total_bill')
plt.show()
```

```python
# Write a program in python to create a histogram for total bill and count
print(f'The histogram becomes:')
sns.histplot(tips_dataset['total_bill'],bins=10,kde=True)
plt.show()
```

```python
sns.kdeplot(tips_dataset['total_bill'],fill=True)
```

```python
# Write a program in python to create a pairplot to see the all the features along with each other
sns.pairplot(tips_dataset)
```

```python
# Write a program in python to find some correlation between the numerical field of the dataset and plot it
corr = tips_dataset[['total_bill','tip','size']].corr()
print(f'The corelation becomes:\n {corr}')
print(f'The distribution becomes:')
sns.heatmap(corr,annot=True,cmap='Accent')
```
