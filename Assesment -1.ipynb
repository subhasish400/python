```python
# Write a program in python to create a function such that it takes the number of people as well as the capacity of the particular lift as input from the user and calculate how many lift rounds are needed to move all people upstairs, given total people and maximum capacity per round. 
def calculate_lift_rounds(n, capacity):
    if capacity <= 0:
        return 0
    rounds = (n + capacity - 1) // capacity
    return rounds
n = eval(input('Enter the capacity of the lift:'))
capacity = eval(input("Enter the capacity of the lift:"))
print(f'The number of rounds to upstair people becomes:\n {calculate_lift_rounds(n,capacity)}')
```

```python
# Write a  program in python to checks if all elements in the list of integers are unique. If all elements are unique, return True; otherwise, return False.
def check_unique(lst):
    a = []
    for x in lst:
        if x in a:
            return False
        a.append(x)
    return True
lst_1 = [1, 2, 3, 4, 5]
print(check_unique(lst_1))
lst_2 = [1, 2, 3, 3, 4, 5]
print(check_unique(lst_2))
```

```python
# Write a program in python to reverse a given list of numbers.
def reverse_lst(lst):
    count = 0
    for x in lst:
        count = count + 1
    left = 0
    right = count - 1
    while left < right:
        temp = lst[left]
        lst[left] = lst[right]
        lst[right] = temp
        left = left + 1 
        right = right - 1 
    return lst 

lst = [1, 2, 3, 4, 5]
print(f'The given list becomes:\n {lst}')
print(f'The reversed list becomes:\n {reverse_lst(lst)}')
```

```python
# Write a program in python to counts and returns the number of even and odd numbers in the given list of integers.
def count_even_odd(lst):
    even_count = 0
    odd_count = 0
    for x in lst:
        if x % 2 == 0:
            even_count = even_count + 1
        else:
            odd_count = odd_count + 1
    return even_count,odd_count
lst = [1, 2, 3, 4, 5]
print(f'The given list of integers becomes:\n {lst}')
print(f'The number of even as well as odd integers becomes:\n{count_even_odd(lst)}')
 
```

```python
# Write a program in python to create a function such that it takes two lists as input argument and if the first list becomes a subset of another one then return true otherwise return false.
def is_subset(lst_1,lst_2):
    for element in lst_1:
        found = False
        for item in lst_2:
            if item == element:
                found = True
                break
        if not found:
            return False
    return True
lst_1 = [1, 2, 3]
lst_2 = [1, 2, 3, 4, 5]
print(is_subset(lst_1,lst_2))
lst_3 = [1, 6]
lst_4 = [1, 2, 3, 4, 5]
print(is_subset(lst_3,lst_4))
```

```python
# Write a program in python to create a function such that it takes a list as input argument and find the maximum difference between two consecutive elements using brute force approach.
def max_consecutive_difference(lst):
    count = 0
    for x in lst:
        count = count + 1 
    max_diff = 0
    for i in range(count-1):
        diff = abs(lst[i] - lst[i+1])
        if diff > max_diff:
            max_diff = diff 
    return max_diff
lst = [1, 7, 3, 10, 5]
print(f'The given list becomes:\n{lst}')
print(f"The maximum difference between it's consecutive elements becomes:\n {max_consecutive_difference(lst)}")
```

```python
# Write a program in python to create a function such that it takes two lists as input from the user and merger them and sort it and display the sorted list.
def merge_two_sorted_lists(list1, list2):
    lst = list1 + list2
    sorted_lst = []
    while lst:
        min_value = lst[0]
        for x in lst:
            if x < min_value:
                min_value = x 
        sorted_lst.append(min_value)
        lst.remove(min_value)
    return sorted_lst
list1 = [1,3,5]
print(f'The first list becomes:\n {list1}')
list2 = [2,4,6]
print(f'The second list becomes:\n {list2}')
print(f'After merging the sorted list becomes:\n {merge_two_sorted_lists(list1, list2)}')
```

```python
# Write a program in python to create a function such that it rotates a given list of numbers by k positions, where k is given by the user.
def rotate_list(lst,k):
    if not lst:
        return []
    count = 0
    for x in lst:
        count = count + 1
    k = k % count
    for i in range(k):
        last = lst.pop()
        lst.insert(0,last)
    return lst
k = eval(input('Enter the time you want to rotate the list:'))
lst =  [1, 2, 3, 4, 5, 6] 
print(f'The given list becomes:\n {lst}')
print(f'The rotated list becomes:\n {rotate_list(lst,k)}')
```

```python
# Write a program in python to create a function such that it takes a sentence as input from the user and print the frequence of word in the entered sentence.
def count_word_frequency(sentence):
    a = sentence.split()
    d = {}
    for x in a:
        if x in d:
            d[x] = d[x] + 1 
        else:
            d[x] = 1
    return d 
sentence = input('enter a sentence:')
print(f'The entered sentence becomes:\n {sentence}')
print(f'The frequency of each word in the sentence becomes:')
for key,value in count_word_frequency(sentence).items():
    print(f'{key} : {value}')
```

```python
# Write a program in python to create a function such that it takes a tuple as input parameter and check weather the entered tuple is palindromic or not.
def is_palindromic_tuple(tup):
    a = list(tup)
    count = 0
    for x in a:
        count = count + 1
    left = 0
    right = count - 1
    while left < right:
        temp = a[left]
        a[left] = a[right]
        a[right] = temp
        left = left + 1
        right = right - 1
    tup1 = tuple(a)
    if tup1 == tup:
        return f'{tup} is a palindromic tuple'
    else:
        return f'{tup} is not a palindromic tuple'
tup = (1, 2, 3, 2, 1)
print(is_palindromic_tuple(tup))
a = (1, 2, 3, 4, 5)
print(is_palindromic_tuple(a))
```

```python
# Write a program in python to create a function that merges multiple dictionaries into a single dictionary. If a key appears in more than one dictionary, sum up their values.
def merge_dicts_with_overlapping_keys(dicts):
    merged = {}
    for d in dicts:
        for key,value in d.items():
            if key in merged:
                merged[key] = merged[key] + value
            else:
                merged[key] = value
    return merged
print(merge_dicts_with_overlapping_keys([{'a': 1, 'b': 2}, {'b': 3, 'c': 4}, {'c': 5, 'd': 6}]))
```
