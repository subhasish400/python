Create two sets: one with the first 5 positive integers and another with the first 5 even integers. Perform and print the results of union, intersection, difference, and symmetric difference operations on these sets.

```python
class Methods:
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def Union(self):
        return self.a.union(self.b)
    def Intersection(self):
        return self.a.intersection(self.b)
    def Difference(self):
        return self.a.difference(self.b)
    def Symmetric_Difference(self):
        return self.a.symmetric_difference(self.b)
a = set([x for x in range(1,6)])
print(f'The first set becomes:\n{a}')
b = set([x for x in range(1,11) if x % 2 == 0])
print(f'The second set becomes:\n{b}')
m1 = Methods(a,b)
print(f'The union of two sets becomes:\n {m1.Union()}')
print(f'The intersection of two sets becomes:\n {m1.Intersection()}')
print(f'The difference of two sets becomes:\n{m1.Difference()}')
print(f'The symmetric difference of two sets becomes:\n {m1.Symmetric_Difference()}')


```

Create two sets: one with the first 5 positive integers and another with the first 3 positive integers. Check if the second set is a subset of the first set and if the first set is a superset of the second set. Print the results.

```python
s1 = {x for x in range(1,6)}
print(f'The first set becomes:\n {s1}')
s2 = {x for x in range(1,4)}
print(f'The second set becomes:\n {s2}')
s2.issubset(s1)
print(f"Is set2 a subset of set1? {s2.issubset(s1)}")
print(f"Is set1 a superset of set2? {s1.issuperset(s2)}")
```
