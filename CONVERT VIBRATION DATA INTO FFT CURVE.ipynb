```python
# Program to convert the vibration data into the FFT curve,
#Here we actually convert the time vs acceleration curve into frequency vs acceleration curve
# here time period (T=1/F) or (F=1/T), where F= natural frequency of vibration
#let here we take the time period T=1 sec and within 1 sec the number of samples N=400
import numpy as np
import matplotlib.pyplot as plt
import scipy.fft as sp
N=400
T=1/400
x=np.linspace(0,N*T,N)
y=3*np.sin(2*np.pi*x)+100*np.sin(50*np.pi*x) #it is the equation of signal
plt.xlabel("Time(sec)")
plt.ylabel("Acceleration")
plt.title("Time vs Acceleration curve")
plt.plot(x,y)
plt.show()
#If we run the following code written above we will get time vs acceleration curve.
```

```python
 #But we have to convert it into the FFT cutrve
import numpy as np
import matplotlib.pyplot as plt
import scipy.fft as sp
N=400
T=1/400
x=np.linspace(0,N*T,N)
y=3*np.sin(2*np.pi*x)+100*np.sin(50*np.pi*x)
yf=(sp.fft(y))
print(yf)
# here the signal is converted into complex numbers.
```

```python
#for plotting we want only real values so,
yf=abs(sp.fft(y))
xf=abs(sp.fftfreq(N,T))
plt.xlabel("Frequency(Hz)")
plt.ylabel("Acceleration")
plt.title("FFT curve(Time vs Acceleration)")
plt.plot(xf,yf,color="Black")
#Here we have to generate some mode on our graph
plt.grid()
plt.show()

# we want the FFT curve to identify the maximum value of acceleration of a vibration at a perticular point.Here we have got the first mode of natural frequency
```

```python
# first observation 
plt.plot(xf,yf, color="Orange")
plt.xlim(0,5)
plt.grid()
```
