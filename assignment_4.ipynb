Create a nested tuple and iterate over the elements, printing each element.

```python
row = eval(input('Enter the number of rows:'))
column = eval(input('Enter the number of columns:'))
a = []
for i in range(row):
    b = []
    for j in range(column):
        n = eval(input('Enter elements:'))
        b.append(n)
    a.append(tuple(b))
T = tuple(a)
print(f'The tuple becomes:\n {T}')
print(f'The each element of nested tuple becomes:')
for sub_tup in T:
    for elem in sub_tup:
        print(elem, end=' ')
```

Create a tuple with duplicate elements. Convert it to a set to remove duplicates and print the resulting set.

```python
T = (1,1,2,2,3,3,3,4,4,5)
print(f'The origional tuple becomes:\n {T}')
a = []
for x in T:
    if x not in a:
        a.append(x)
print(f'After removing the duplicates the tuple becomes:\n {tuple(a)}')
# Alternative way:
b = set(T)
print(f'The resulting set becomes:\n {b}')
```

Write functions that take a tuple and return the minimum, maximum, and sum of the elements. Print the results for a sample tuple.

```python
class Operations:
    def __init__(self, tup):
        self.tup = tup
    def maximum(self):
        max = self.tup[0]
        for x in self.tup:
            if x > max:
                max = x
        return max
    def minimum(self):
        min = self.tup[0]
        for x in self.tup:
            if x < min:
                min = x
        return min
    def total(self):
        total = 0
        for x in self.tup:
            total = total + x
        return total
O1 = Operations([x for x in range(1,6)])
print(f'The maximum value in the tuple becomes:\n {O1.maximum()}')
print(f'The minimum value in the tuple becomes:\n {O1.minimum()}')
print(f'The total value of the tuple becomes:\n {O1.total()}')
```
